### 1. Some todos os números de 1 até 100

```python
sum = 0
for i in range(1, 101):
    sum += i
    
print(f"Soma: {sum}")
```

### 2. Some todos os números pares de 1 até 100

```python
# Usando divisão com resto zero, somando apenas os pares
sum = 0
for i in range(1, 101):
  if i % 2 == 0:
    sum += i
    
print(f"Soma Pares: {sum}")
```

### 3. Some todos os números ímpares de 1 até 100

```python
# Usando range com passo 2, começando de 1 somarei apenas os impares
sum = 0
for i in range(1, 101, 2):
    sum += i
    
print(f"Soma Impares: {sum}")
```

### 4. Dada a lista abaixo, some todos os elementos inteiros positivos

```python
lista = [-2, 2, 'celular', 10, -99, False]
lista

sum = 0
for i in lista:
  if isinstance(i, int) and i >= 0 and i != False and i != True:
    print(f"i: {i}")
    sum += i
    
print(f"Soma: {sum}")
```

### 5. Dada a String

```python
string = 'pYthOn É A mELhoR LiNguaGEM De PRoGRaMaçãO'
string
```

Formate-a com as primeiras letras maiúsculas:

```python
string_formatada = string.title()
print(f"{string_formatada}")
```

Com todas as letras minúsculas

```python
for i in string:
  lower = i.lower() #transformei todas em minusculo, mesmo as que ja estavam
  print(f"{lower}", end='')
```

Com todas as letras maiúsculas

```python
for i in string:
  upper = i.upper() #transformei todas em maiusculo, mesmo as que ja estavam
  print(f"{upper}", end='')
```
